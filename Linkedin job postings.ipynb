# Web Scraping of Job Posting from LinkedIn website:
Here we have tried to web scrap from LinkedIn website for 2 locations USA and Canada with different job search keywords like Google cloud, AWS cloud, Azure Cloud, Google for education, G Suite(google workspace) and others.

We have changed the input as text input for different searches and locations.

1. Programming Language used: Python
2. Libraries used: Beautiful soup, requests, Pandas
3. Jupyter Notebooks

## Results:
We have extracted 24 results for all the search keywords (Google cloud, AWS cloud and others) for USA and canada locations, saved in results_LinkedIn.csv file.

```python
#import chrome webdriver
# ! pip install selenium
# ! pip install bs4
#! pip install lxml
from selenium import webdriver
from bs4 import BeautifulSoup as soup
import pandas as pd
import requests
```

```python
x = input("Enter job profile: ").split(" ")
x = '-'.join(x)
y = input("location: ")
#searching the user input of monster
URL = "https://www.linkedin.com/jobs/search?keywords="+x+"&location="+y
page = requests.get(URL)
scr = soup(page.content, 'lxml')
print(URL)
```

```python
#Scrapping for different information regarding each job such as profile, company name, location for job and links to apply on.
profile=[]
for header in scr.findAll('h3', {'class':'result-card__title job-result-card__title'}):
    profile.append(header.text.strip())
company=[]
for name in scr.findAll('a', {'class':'result-card__subtitle-link job-result-card__subtitle-link'}):
    company.append(name.text.strip())
# links=[]
# for link in scr.findAll('h2', {'class':'title'}):
#     links.append(link.a['href'])
location=[]
for locs in scr.findAll('span', {'class':'job-result-card__location'}):
    location.append(locs.text.strip())
```

```python
# making a dictionary of the scrapped data.
d1 = {'Profile': profile, 'Company Name': company,'Location': location}
# comverting Data into dataframe.
df = pd.DataFrame.from_dict(d1)
#df.head()
#Exporting scrapped data into CSV file
df.to_csv(r'results_LinkedIn.csv', index = False)#'demo\jobs is temporary saving location.'
# #display Data.
df
```
